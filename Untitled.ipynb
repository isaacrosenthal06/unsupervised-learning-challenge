```python
import pandas as pd
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
myopia = "Resources/myopia.csv"
myopia_df = pd.read_csv(myopia)
myopia_df.head()
```

```python
clean_df = myopia_df.drop(columns= ['MYOPIC'])
clean_df.head()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clean_df[['AGE', 'SPHEQ', 'AL', 
                                                'ACD', 'LT', 'VCD', 'SPORTHR', 'READHR',
                                                  'COMPHR', 'STUDYHR', 'TVHR', 'DIOPTERHR', 
                                                  'MOMMY', 'DADMY']])

scaled_df = pd.DataFrame(scaled_data, columns= ['AGE', 'SPHEQ', 'AL', 
                                                'ACD', 'LT', 'VCD', 'SPORTHR', 'READHR',
                                                  'COMPHR', 'STUDYHR', 'TVHR', 'DIOPTERHR', 
                                                  'MOMMY', 'DADMY'])
scaled_df.head()
```

```python
pca = PCA(n_components = .95)
myopia_pca = pca.fit_transform(scaled_df)
```

```python
pca.explained_variance_ratio_
```

```python
myopia_pca_df = pd.DataFrame(data=myopia_pca, columns=["PCA1", "PCA2", "PCA3", "PCA4",
                                                      "PCA5", "PCA6", "PCA7", "PCA8",
                                                      "PCA9", "PCA10", "PCA11"])
myopia_pca_df.head()
```

```python
tsne = TSNE(learning_rate=90)
tsne_features = tsne.fit_transform(myopia_pca_df)
tsne_features.shape
```

```python
myopia_pca_df['x'] = tsne_features[:,0]

myopia_pca_df['y'] = tsne_features[:,1]

plt.scatter(myopia_pca_df['x'], myopia_pca_df['y'])
plt.show()
```

```python
labels = myopia_df['MYOPIC']
plt.scatter(myopia_pca_df['x'], myopia_pca_df['y'], c=labels)
plt.show()
```

```python
model = KMeans(n_clusters=10, random_state=5)
model.fit(myopia_pca_df)

predictions = model.predict(myopia_pca_df)

myopia_pca_df["class"] = model.labels_
myopia_pca_df.head()
```

```python
inertia = []
k = list(range(1, 12))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(myopia_pca_df)
    inertia.append(km.inertia_)

elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters=5, random_state=5)
model.fit(myopia_pca_df)
```
